Data Cleaning

```python
data.isnull().sum()
```

```python
print(data.dtypes)
```

```python
data.duplicated().any()
```

```python
data.replace( {'Payment': {'Cash': 1, 'Credit card': 2, 'Ewallet': 3}}, inplace=True)
```

```python
data.replace( {'Gender': {'Female': 0, 'Male':1}}, inplace=True)
```

```python
data.head()
```

```python
from scipy import stats

z_scores = np.abs(stats.zscore(data.select_dtypes(include=[np.number])))
outliers = data[(z_scores > 3).any(axis=1)]
print(outliers)
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=data.select_dtypes(include=[np.number]))
plt.show()
```

```python
sales_per_branch = data.groupby('Branch')['Total'].sum().reset_index()
sales_per_branch.columns = ['Branch', 'Total_Sales']
print(sales_per_branch)
```

```python
avg_gross_income = data.groupby('Product line')['gross income'].mean().reset_index()
avg_gross_income.columns = ['Product_Line', 'Avg_Gross_Income']
print(avg_gross_income)
```

```python
gender_segment = data.groupby('Gender')['Invoice ID'].count().reset_index()
gender_segment.columns = ['Gender', 'Customer_Count']
print(gender_segment)
```

```python
# Bar plot for total sales per branch
plt.figure(figsize=(7, 4))
sns.barplot(x='Branch', y='Total_Sales', data=sales_per_branch)
plt.title('Total Sales per Branch')
plt.show()

# Bar plot for average gross income by product line
plt.figure(figsize=(12, 4))
sns.barplot(x='Product_Line', y='Avg_Gross_Income', data=avg_gross_income)
plt.title('Average Gross Income by Product Line')
plt.show()
```

Exploratory Data Analysis

```python
# Sales Trend Analysis

# Convert your 'Date' column to datetime, with dayfirst=True for European-style dates
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)


# Plot sales trends over time
plt.figure(figsize=(7, 4))
data.groupby(data['Date'].dt.to_period('M'))['Total'].sum().plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()

# Seasonality (Weekly Sales Trend)
plt.figure(figsize=(7, 4))
data.groupby(data['Date'].dt.to_period('W'))['Total'].sum().plot()
plt.title('Weekly Sales Trend')
plt.xlabel('Week')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Product Line Analysis
# Sales by product line across branches
plt.figure(figsize=(7, 4))
sns.barplot(x='Product line', y='Total', hue='Branch', data=data, estimator=sum)
plt.title('Sales by Product Line Across Branches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Payment Methods Analysis
plt.figure(figsize=(7, 4))
sns.countplot(x='Payment', data=data)
plt.title('Distribution of Payment Methods')
plt.show()

# Sales by Payment Method
plt.figure(figsize=(7, 4))
sns.barplot(x='Payment', y='Total', data=data)
plt.title('Sales by Payment Method')
plt.show()

# Payment Method vs. Customer Satisfaction (assuming you have a 'Rating' column)
plt.figure(figsize=(7, 4))
sns.barplot(x='Payment', y='Rating', data=data)
plt.title('Customer Satisfaction by Payment Method')
plt.show()
```

Performance Analysis

```python
# Branch Performance
# Total Sales by City
sales_by_city = data.groupby('City')['Total'].sum().reset_index()
plt.figure(figsize=(7, 4))
sns.barplot(x='City', y='Total', data=sales_by_city)
plt.title('Total Sales by City')
plt.show()
```

```python
# Sales Performance by Customer Type
# Total Sales by Customer Type
sales_by_customer_type = data.groupby('Customer type')['Total'].sum().reset_index()
plt.figure(figsize=(7, 4))
sns.barplot(x='Customer type', y='Total', data=sales_by_customer_type)
plt.title('Total Sales by Customer Type')
plt.show()
```

```python
# Product Line Profitability
# Gross Income by Product Line
gross_income_by_product_line = data.groupby('Product line')['gross income'].sum().reset_index()
plt.figure(figsize=(7, 4))
sns.barplot(x='Product line', y='gross income', data=gross_income_by_product_line)
plt.title('Gross Income by Product Line')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gross Income and Margins
# Add a column for gross margin percentage
data['Gross Margin %'] = (data['gross income'] / data['Total']) * 100

# Average Gross Margin by Product Line
gross_margin_by_product_line = data.groupby('Product line')['Gross Margin %'].mean().reset_index()
plt.figure(figsize=(7, 4))
sns.barplot(x='Product line', y='Gross Margin %', data=gross_margin_by_product_line)
plt.title('Average Gross Margin by Product Line')
plt.xticks(rotation=45)
plt.show()

# Overall Gross Income and Gross Margin
total_gross_income = data['gross income'].sum()
average_gross_margin = data['Gross Margin %'].mean()

print(f"Total Gross Income: {total_gross_income}")
print(f"Average Gross Margin Percentage: {average_gross_margin:.2f}%")
```

Customer Satisfaction Analysis

```python
# Customer Satisfaction by Branch
# Average customer rating by branch
avg_rating_by_branch = data.groupby('Branch')['Rating'].mean().reset_index()

# Plot the average customer rating by branch
plt.figure(figsize=(7, 4))
sns.barplot(x='Branch', y='Rating', data=avg_rating_by_branch)
plt.title('Average Customer Rating by Branch')
plt.show()
```

```python
# Customer Satisfaction by Product Line
# Average customer rating by product line
avg_rating_by_product_line = data.groupby('Product line')['Rating'].mean().reset_index()

# Plot the average customer rating by product line
plt.figure(figsize=(7, 4))
sns.barplot(x='Product line', y='Rating', data=avg_rating_by_product_line)
plt.title('Average Customer Rating by Product Line')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Satisfaction by Payment Method
# Average customer rating by payment method
avg_rating_by_payment = data.groupby('Payment')['Rating'].mean().reset_index()

# Plot the average customer rating by payment method
plt.figure(figsize=(7, 4))
sns.barplot(x='Payment', y='Rating', data=avg_rating_by_payment)
plt.title('Average Customer Rating by Payment Method')
plt.show()
```

```python
# Factors Impacting Customer Ratings
# Correlation matrix
# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=[np.number])
# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the correlation matrix
plt.figure(figsize=(7, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```
